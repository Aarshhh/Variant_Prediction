```python
test_data = pd.read_csv("Test.csv")
test_data
```

```python
train_data = pd.read_csv("Train.csv")
train_data
```

```python
X_train = pd.DataFrame(train_data.drop(['Variant'],axis=1))
Y_train = pd.DataFrame(train_data['Variant'])
test_data['Variant'] = train_data['Variant']
```

```python
X_test = pd.DataFrame(test_data.drop(['Variant'],axis=1))
Y_test = pd.DataFrame(test_data['Variant'])
```

```python
a = pd.get_dummies(X_test['family history'])
```

```python
b = pd.get_dummies(X_test['family history'])
```

```python
X_test[['New Column','Second Column']] = a
X_train[['New Column','Second Column']] = b
mean_test = X_test['New Column'].mean()
mean_test2 = X_test['Second Column'].mean()
X_test['New Column'].fillna(value=mean_test, inplace=True)
X_test['Second Column'].fillna(value=mean_test2,inplace=True)
mean_train = X_train['New Column'].mean()
mean_train2 = X_train['Second Column'].mean()
X_train['New Column'].fillna(value=mean_train,inplace=True)
X_train['Second Column'].fillna(value=mean_train2,inplace=True)
```

```python
c = pd.get_dummies(X_train['Region'])
d = pd.get_dummies(X_test['Region'])
X_train[['Third','Fourth','Five','Six']] = d
X_test[['Third','Fourth','Five','Six']] = c
X_train.drop('Region',axis=1,inplace=True)
X_test.drop('Region',axis=1,inplace=True)
```

```python
X_train.drop('Pre-existing symptoms',axis=1,inplace=True)
X_test.drop('Pre-existing symptoms',axis=1,inplace=True)
```

```python
g = pd.get_dummies(X_train['family history'])
f = pd.get_dummies(X_test['family history'])
```

```python
X_train[['Seven','Eight']] = g
X_test[['Seven','Eight']] = f
X_train.drop('family history',axis=1,inplace=True)
X_test.drop('family history',axis=1,inplace=True)
```

```python
X_train.fillna(0,inplace=True)
X_test.fillna(0,inplace=True)
Y_test.fillna(0,inplace=True)
Y_train.fillna(0,inplace=True)
X_train.apply(pd.isnull).sum()/X_train.shape[0]
```

```python
from numpy import loadtxt
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
```

```python
# fit model no training data
model = XGBClassifier()
model.fit(X_train, Y_train)
```

```python
print(model)
```

```python
# make predictions for test data
Y_pred = model.predict(X_test)
predictions = [(value) for value in Y_pred]

df = pd.DataFrame(test_data['ID'])
df['Variant'] = Y_pred
df.to_csv("GDXBoost.csv")
```

```python
accuracy = accuracy_score(Y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
